```python
#First task: fibonacci
def fib(n):
    if n <= 1:
        return n
    else:
        return (fib(n - 1) + fib(n - 2))
    
n = int(input("Quantity of first fibonacci elements:"))
for i in range(n):
    print (fib(i))
```

```python
#second task

def fact(x):
    if x <= 1:
        return 1
    else:
        return (x * fact(x - 1))
print (fact(5))
```

```python
#third task

def mass_index(height, weight):
    return weight/ height**2

height = float(input("Type in height:"))
weight = float(input("type in weight:"))
mass_index(height, weight)
```

```python
#fifth task

import math
def sum_of_cubes(n):
    s = 0
    for i in range(n + 1):
        s = s + math.pow(i, 3)
    return s

sum_of_cubes(4)
```
